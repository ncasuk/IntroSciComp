<H1>Two lines with axes and a title</H1>

```python
import matplotlib.pyplot as plt
times = [0, 0.25, 0.5, 0.75]
plt.plot(times, [0,0.5,1,1.2], 'g--')
plt.plot(times, [1, 2, 3, 4],  'r')
plt.title('Concentration of Chlorine vs Time')
plt.ylabel('Concentration (%)')
plt.xlabel('Time (s)')
plt.show()
```

```python
# Add a legend

import matplotlib.pyplot as plt
times = [0, 0.25, 0.5, 0.75]
plt.plot(times, [0,0.5,1,1.2], 'g--', label = "Some data")
plt.plot(times, [1, 2, 3, 4],  'r', label = "Other data")
plt.title('Concentration of Chlorine vs Time')
plt.ylabel('Concentration (%)')
plt.xlabel('Time (s)')
plt.legend()
plt.show()
```

```python
# Histogram plot

import numpy as np
import matplotlib.pyplot as plt
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
n, bins, patches = plt.hist(x,50,density=1,color='g',
edgecolor='k')
plt.axis([40, 160, 0, 0.03])
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.show()
```

```python
# Multiple plots – data and subplot function

import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

The "subplot" function is defined as: 
subplot(nrows, ncols, plot_number)
Here we define: 2 rows, 1 column.

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-', markeredgecolor='k')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.show()
```

```python
# Multiple axes on one plot

import numpy as np
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
ax1.plot(t, s1, 'b-')
ax1.set_xlabel('time (s)')

# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('exp', color='b')
for tl in ax1.get_yticklabels():
    tl.set_color('b')
    
ax2 = ax1.twinx()
s2 = np.sin(2*np.pi*t)
ax2.plot(t, s2, 'r.')
ax2.set_ylabel('sin', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')

plt.show()
```

```python
# Contour plot – prepare data

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)

# Make field to contour
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
```

```python
# What shape are X, Y, Z?
print(np.shape(X), np.shape(Y), np.shape(Z))
```

```python
# Contour plot – default colours 

cs = plt.contour(X, Y, Z)
plt.clabel(cs, inline=1, fontsize=10)
plt.title('Simplest default with labels')
plt.show()
```

```python
# Contour plot – with negative values
# This time negative contours will be dashed by default

CS = plt.contour(X, Y, Z, 6, colors='k')
plt.clabel(CS, fontsize=9, inline=1)
plt.title('Single color - negative contours dashed')
plt.show()
```

```python
# Contour plot - set negative line style
# Override negative contours – use solid lines

matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
CS = plt.contour(X, Y, Z, 6, colors='k')
plt.clabel(CS, fontsize=9, inline=1)
plt.title('Single color - negative contours solid')
plt.show()
```

```python
# Contour plot – specify colours

CS = plt.contour(X, Y, Z, 6, linewidths=np.arange(1, 8, 1.5), colors=('r', 'green', 'blue', 'yellow', 'cyan', 'wheat'))
plt.clabel(CS, fontsize=9, inline=1)
plt.title('Crazy lines')
```

<H1>Introducing cartopy</H1>

<B>cartopy</B> is a Python package designed for geospatial data processing in order to produce maps and other geospatial data analyses.<P>
    
<B>cartopy</B> does not do any plotting on its own, but provides the facilities to transform coordinates to one of 33 different map projections.<P>
    
<B>matplotlib</B> is then used to plot contours, images, vectors, lines or points in the transformed coordinates. Shoreline, river and political boundary datasets are provided within cartopy, along with methods for plotting them

<H1>Making a contour plot on a map</H1>

In this example we will extract the air temperature field from the gdata.nc netCDF file and plot on a map to make this plot




```python
# Plotting data on maps – code 1/3

# Import libraries
import matplotlib.pyplot as plt
import cartopy
import cartopy.crs as ccrs
import cartopy.util as cartopy_util
import numpy as np
from netCDF4 import Dataset as ncfile

# Read in the data
nc = ncfile('gdata.nc')
lons=nc.variables['lon'][:]
lats=nc.variables['lat'][:]
field=nc.variables['temp'][0,:,:]


```

```python
# Let's look at the longitude array
print(lons)
```

```python
# Plotting data on maps – code 2/3
# Add a cyclic point so that the data wraps in longitude
#lonrange = np.max(lons)-np.min(lons)
#if lonrange <= 360:
#    field, lons = cartopy_util.add_cyclic_point(field, lons)

# Set the plot limits
lonmin=-180
lonmax=180
latmin=-90
latmax=90
lon_0=(lonmax-lonmin)/2.0+lonmin
```

```python
# Plotting data on maps – code 3/3

# Open a figure
plot=plt.figure(figsize=(11, 8))

# Set the map
proj=ccrs.PlateCarree(central_longitude=lon_0)
mymap = plot.add_subplot(1,1,1, projection=proj)
mymap.set_extent((lonmin, lonmax, latmin, latmax), crs=proj)

# Set the cotour levels
clevs=np.arange(-32, 32, 4) # levels

# Contour the data - colour fill, lines and labels
cf = mymap.contourf(lons, lats, field, clevs,extend='both', \
     transform=ccrs.PlateCarree(), cmap='bwr')

cl = mymap.contour(lons, lats, field, clevs, colors='k', \
     linestyles='solid', transform=ccrs.PlateCarree())

plt.clabel(cs, fmt = '%d', colors = 'k', fontsize=11) 

# Add a colour bar
plot.colorbar(cf, orientation='horizontal', aspect=75, pad=0.08, ticks=clevs)

# Add some axes
mymap.set_xticks(np.arange(-180, 181, 60), crs=proj)
mymap.set_yticks(np.arange(-90, 91, 30))

# Add the coastlines
mymap.coastlines(resolution='110m')

# Set the plot title
title = plt.title('Temperature in degrees Celsius', y=1.03)
```
